```python
from scipy.io import loadmat
import pandas as pd
import numpy as np
alpha = loadmat('alpha2.mat')
y_2 = alpha['MeanNums'].flatten()
y_1 = alpha['MeanSizes'].flatten()
x_ = alpha['x'].flatten()
df = pd.DataFrame({'alphas':x_,
                   'MeanSizes':y_1,
                    'MeanNums':y_2})
```

```python
df
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
plt.style.use('grayscale')
plt.rcParams.update({"figure.facecolor":(0,0,0,0)})
#sns.set_style("darkgrid")
fig,ax = plt.subplots(figsize=(7,3),dpi=150)
s = ax.scatter(df['MeanSizes'],df['MeanNums'],c=df['alphas'],edgecolor = 'k',cmap='binary')
ax.set_xlabel(r'Mean size of events', fontsize=12)
ax.set_ylabel(r'Mean amount of events', fontsize=12)
fig.colorbar(s,ax=ax,label=r'$value\ of\ \alpha$')
ax.grid(True)
fig.tight_layout()
fig.savefig('./alpha_reviewer2.png',bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set_style("darkgrid")
from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1)) 


#fig = plt.figure(figsize=(15,8))
fig,ax = plt.subplots(figsize=(8,3),dpi=300)
bar = sns.barplot(data=df, x='alphas',y='MeanSizes',ax=ax,color='gray',label='event size') #cornflowerblue
plt.legend(loc="upper left", bbox_to_anchor=(0.7,0.6))
ax.set_xlabel('Alphas')
ax.set_ylabel('Average Size of events (in pixel)')

ax2 = ax.twinx()
line = sns.lineplot(data=df['MeanNums'],color='k',marker="o",ax=ax2,label='event amount')#color='salmon'
ax2.grid(False)
ax2.set_ylabel('Average amount of events')
ax2.yaxis.set_major_formatter(formatter) 
#ax2 = sns.lineplot(data=df,x='alphas',y='MeanNums',color='salmon',markers="o",ax=ax2)
plt.legend(loc="upper left", bbox_to_anchor=(0.7,0.5))

plt.savefig('./alpha_new.png',bbox_inches='tight')
plt.show()
```
