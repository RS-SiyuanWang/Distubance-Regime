```python
def intensity(beta,area,intercept=0.22684):
    import numpy as np
    return beta*np.log10(100*area)+intercept
sizes = []
import numpy as np
for i in np.arange(20):
    size = 2**i
    sizes.append(size)
betas = [*np.arange(0.03,0.10,0.01), *np.arange(0.15,0.30,0.05), 0.30,0.40,0.50]
betas = list(np.around(np.array(betas),2))
import pandas as pd
colnames = [str(x) for x in betas]
df = pd.DataFrame(columns=colnames)
for beta in betas:
    Is = []
    for j in sizes:
        I = intensity(beta,j)
        Is.append(I)
    df[str(beta)]=Is
df.insert(loc = 0, column='areas',value = sizes)
```
